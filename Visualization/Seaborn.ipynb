# Визуализация отношений в данных

## Функция lineplot

Такой тип графика чаще всего используется для визуализации временных рядов и зависимостей между переменными, имеющими непрерывный характер.

```python
sns.set_style("whitegrid")
```

```python
date = pd.date_range(start="2021-09-01", freq="D", periods=8)
day = [23, 17, 17, 16, 15, 14, 17, 20]
night = [19, 11, 16, 11, 10, 10, 11, 16]
df = pd.DataFrame({'date':date, 'day_temperature': day, 'night_temperature': night})
```

```python
date
```

```python
sns.lineplot(data=df);
```

```python
sns.lineplot(x="night_temperature", y="day_temperature", data=df);
```

```python
sns.lineplot(x="night_temperature", y="day_temperature", data=df, sort=False);
```

```python
sns.set_style("darkgrid")
data = sns.load_dataset("mpg")
data
```

```python
sns.lineplot(x='model_year', y='horsepower', hue="origin", data=data);
```

## Функция scatterplot

Построим точечный график зависимости дальности пробега (displacement) автомобиля от расхода топлива (mpg):

```python
sns.scatterplot(x='mpg', y='displacement', data=data);
```

```python
sns.scatterplot(x='mpg', y='displacement', hue='origin', data=data);
```

## Функция relplot

В отличие от предыдущих функций, эта функция предоставляет возможность настраивать не только внешний вид графика, но и фигуру — подложку, на которой располагаются все графические компоненты.

```python
sns.relplot(x='mpg', y='displacement', kind='scatter', data=data);
```

```python
sns.relplot(x='mpg', y='displacement', kind='scatter', hue='origin', col='origin', data=data);
```

# Визуализации распределения категориальных данных

## Функция stripplot

Функция строит точечную диаграмму и по функционалу похожа на функцию scatterplot

```python
sns.stripplot(x='origin', y='mpg', data=data);
```

## Функция swarmplot

Функция аналогична функции stripplot, но отображаемые точки на диаграмме не перекрываются, это позволяет делать выводы о преобладании тех или иных значений в наборах данных по их визуальному распределению.

```python
sns.swarmplot(x='origin', y='mpg', data=data);
```

```python
from numpy import median
sns.lmplot(x="horsepower", y="displacement", data=data, x_estimator=median);
```

## Функция residplot

Функция отображает отклонения элементов исходного набора данных от регрессионной модели, построенной по ним, в виде диаграммы рассеяния. Каждая точка такой диаграммы — это разность между значением элемента исходного набора и значением, которое выдаст модель регрессии в этой точке.

```python
sns.residplot(x="horsepower", y="displacement", data=data);
```

## Функция lmplot

Функция полный аналог функции regplot с возможностью управлять компоновкой полей с графиками на подложке как для функции relplot.

```python
sns.lmplot(x="horsepower", y="displacement", hue="origin", col="origin", data=data);
```
