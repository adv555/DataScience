# Узор Серпинского

Реализуйте программу которая будет строить [узор Серпинского](https://ru.wikipedia.org/wiki/%D0%A2%D1%80%D0%B5%D1%83%D0%B3%D0%BE%D0%BB%D1%8C%D0%BD%D0%B8%D0%BA_%D0%A1%D0%B5%D1%80%D0%BF%D0%B8%D0%BD%D1%81%D0%BA%D0%BE%D0%B3%D0%BE#%D0%98%D1%82%D0%B5%D1%80%D0%B0%D1%82%D0%B8%D0%B2%D0%BD%D1%8B%D0%B9_%D0%BC%D0%B5%D1%82%D0%BE%D0%B4). История исследования этого узора достаточно длинная, а сам узор представляет особый интерес в такой области, как геометрия фракталов. Узор Серпинского является случайным объектом, определенным рекурсивно, но в пределе его форма стремится к детерминированному объекту.

```python
def sierpinski(a, b, c, level):

    if level == 0:
        plt.scatter([a[0], b[0], c[0]], [a[1], b[1], c[1]], alpha=0.5)
        # plt.plot([a[0], b[0], c[0], a[0]], [a[1], b[1], c[1], a[1]], 'k')
    else:
        x = (a[0] + b[0]) / 2
        y = (a[1] + b[1]) / 2

        x1 = (a[0] + c[0]) / 2
        y1 = (a[1] + c[1]) / 2

        x2 = (b[0] + c[0]) / 2
        y2 = (b[1] + c[1]) / 2

        sierpinski(a, (x, y), (x1, y1), level - 1)
        sierpinski((x, y), b, (x2, y2), level - 1)
        sierpinski((x1, y1), (x2, y2), c, level - 1)

font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 16,
        }

sierpinski((-200, -100), (0, 200), (200, -100), 5)
plt.title('Sierpinski triangle', fontdict=font, x=0.5, y=1.1)
plt.subplots_adjust(top=1.5)
plt.show()
```
