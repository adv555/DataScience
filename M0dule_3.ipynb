```python
import matplotlib.pyplot as plt
import pandas as pd

date = pd.date_range(start='2022-01-01', freq='D', periods=8)
```

```python
plt.rcParams['figure.figsize'] = [10,8]
plt.plot(date,[-1,-2,-5,0,-4,-3,0,-4])
```

```python
# matplotlib.text.Text
plt.xlabel('Дата', fontsize='small', color='midnightblue')
plt.ylabel('Температура', fontsize='small', color='midnightblue')

plt.title('Средняя температура по городу')
plt.text(date[0], -5, 'самый холодный день')
```

```python
date = pd.date_range(start='2022-01-01', freq='D', periods=8)
plt.plot(date,[-1,-2,-5,0,-4,-3,0,-4], label='meen date temperature')
plt.xlabel('Дата', fontsize='small', color='midnightblue')
plt.ylabel('Температура', fontsize='small', color='midnightblue')
plt.title('Средняя температура по городу')
plt.text(date[0], -5, 'самый холодный день')
plt.legend()
plt.show()
```

```python
date = pd.date_range(start='2022-01-01', freq='D', periods=8)
fig, axs = plt.subplots()

axs.plot(date,[-5,-3,-8,0,-5,-6,0,-7], label='meen night temperature')
axs.plot(date,[-1,-2,-5,0,-4,-3,0,-4], label='meen date temperature')

plt.xlabel('Дата', fontsize='small', color='midnightblue')
plt.ylabel('Температура', fontsize='small', color='midnightblue')
plt.title('Средняя температура по городу')
plt.text(date[2], -5.2, 'самый холодный день')
plt.text(date[2], -8.2, 'самая холодная ночь')
plt.legend()
plt.show()
```

```python
date = pd.date_range(start='2022-01-01', freq='D', periods=8)
fig, axs = plt.subplots(2,1)

axs[0].plot(date,[-5,-3,-8,0,-5,-6,0,-7], label='meen night temperature')
axs[1].plot(date,[-1,-2,-5,0,-4,-3,0,-4], label='meen date temperature')

axs[0].set_title('Ночь', fontsize=10)
axs[1].set_title('День', fontsize=10)

fig.suptitle('Средняя температура по городу')

plt.show()
```

```python
date = pd.date_range(start='2022-01-01', freq='D', periods=8)
plt.plot(
    date,
    [-5,-3,-8,0,-5,-6,0,-7],
    label='mean date temperature',
    linestyle = "--",
    color = "#ff5733"
)

plt.plot(
    date,
    [-1,-2,-5,0,-4,-3,0,-4],
    label='mean night temperature',
    linestyle = ":",
    color = "#061358"
)

plt.xlabel('Дата', fontsize='small', color='midnightblue')
plt.ylabel('Температура', fontsize='small', color='midnightblue')
plt.title('Средняя температура по городу')
plt.legend()
plt.show()
```

```python
date = pd.date_range(start='2022-01-01', freq='D', periods=8)
plt.plot(
    date,
    [-5,-3,-8,0,-5,-6,0,-7],
    label='mean date temperature',
    linestyle = "--",
    color = "#ff5733",
    linewidth = 2,
    marker ='D',
)

plt.plot(
    date,
    [-1,-2,-5,0,-4,-3,0,-4],
    label='mean night temperature',
    linestyle = ":",
    color = "#061358",
    linewidth = 2,
    marker ='^',
)

# plt.ylim(-8,0)
plt.xlabel('Дата', fontsize='small', color='midnightblue')
plt.ylabel('Температура', fontsize='small', color='midnightblue')
plt.title('Средняя температура по городу')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np

plt.rcdefaults()
fif, axs = plt.subplots()
people = ('John', 'Bill', 'Tom')
y_pos = np.arange(len(people))
perfomance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

axs.barh(y_pos, perfomance, xerr=error, align='center')
axs.set_yticks(y_pos, labels=people)
axs.invert_yaxis()

axs.set_xlabel('Perfomance')
axs.set_title('Заголовок')
plt.show()
```

```python
labels = ['Junior','Senior','Engineer','SysAdmin']
data = [64,32,100,2]
explode = [0.15,0,0,0]

plt.pie(
    data,
    labels=labels,
    shadow=True,
    explode=explode,
    autopct='%.2f%%',
    pctdistance=1.15,
    labeldistance= 1.35
)

plt.show()
```

```python
theta = np.linspace(0, 2.0*np.pi, 1000)
r = np.sin(6*theta)
plt.polar(theta,r)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

theta_max = 8*np.pi
n=1000
theta = np.linspace(0,theta_max,n)
x=theta
y=np.cos(theta)
z=np.sin(theta)

ax.plot(x,y,z, 'g');
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

x=[5,10,15,20]
y=[10,0,5,25]
z=[0,10,5,15]
s=[150,130,30,160]

ax.scatter(x,y,z,s=s)

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

grid=np.arange(-10,10,0.5)
x,y = np.meshgrid(grid,grid)
z = x**2*y**2+2

ax.plot_wireframe(x,y,z)

plt.show()
```
